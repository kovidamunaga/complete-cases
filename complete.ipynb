```r
######## coursera assignment###### 
##### mean function######
pollutantmean<- function(directory,pollutant,id=1:332){
  filelist<-list.files(path=directory,full.names = TRUE,pattern=".csv")
  values<-numeric()
  
  for(i in id){
    data<-read.csv(filelist[i])
    values<-c(values,na.omit(data[[pollutant]]))
  }
  mean(values)
}
```

```r

#######complete function###

complete<-function(directory,id=1:332){
  file<-list.files(directory,pattern = ".csv",full.names = TRUE)
  obs<-numeric()
  
  for(i in id){
    data<-(sum(complete.cases(read.csv(file[i]))))
    obs<-c(obs,data)
    
  }
  data.frame(id,obs)
}
```
